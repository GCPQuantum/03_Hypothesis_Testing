# Hypothesis Testing

```python
x_val=np.arange(0,11)
y_val=binom.pmf(n=10,k=x_val,p=0.5)
sns.barplot(x=x_val,y=y_val)
```

```python
h_val=[]
for i in range(1000):
    size=100
    h_val.append(pd.value_counts(np.random.choice(["H","T"],size=size))["H"])
sns.kdeplot(h_val)
```

```python
h_val=[]
for i in range(1000):
    size=1000
    h_val.append(pd.value_counts(np.random.choice(["H","T"],p=[0.7,0.3],size=size))["H"])
sns.kdeplot(h_val)
```
